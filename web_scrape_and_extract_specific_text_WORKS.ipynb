# WEB SCRAPING THE HINDU and LINKEDIN

```python
import requests
from bs4 import BeautifulSoup
import openai

# Function to scrape content from a webpage
def scrape_content(url):
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        return soup.get_text()
    else:
        print(f"Failed to retrieve {url}")
        return ""

# List of URLs to scrape
urls = [
    "https://www.thehindu.com/",
    "https://www.linkedin.com/in/kunal-kumar-abb66528b/",
]

# Scraping the content of each webpage
webpages_content = []
for url in urls:
    content = scrape_content(url)
    webpages_content.append(content)
    print(f"Scraped content from {url}\n")

# Show all the scraped contents
for i, content in enumerate(webpages_content):
    print(f"Content from URL {i+1}:\n{content}\n")
```
