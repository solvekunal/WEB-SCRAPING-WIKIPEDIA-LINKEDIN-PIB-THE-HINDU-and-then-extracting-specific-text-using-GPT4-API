```python
import requests
from bs4 import BeautifulSoup
import openai
```

```python
# Function to scrape content from a webpage
def scrape_content(url):
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        return soup.get_text()
    else:
        print(f"Failed to retrieve {url}")
        return ""
```

```python
# List of URLs to scrape
urls = [
    "https://pib.gov.in/PressReleasePage.aspx?PRID=2031358",
    "https://en.wikipedia.org/wiki/Ashoka",
]
```

```python
# Scraping the content of each webpage
webpages_content = []
for url in urls:
    content = scrape_content(url)
    webpages_content.append(content)
    print(f"Scraped content from {url}\n")

# Show all the scraped contents
for i, content in enumerate(webpages_content):
    print(f"Content from URL {i+1}:\n{content}\n")

# Function to use GPT-4 to extract specific text
def extract_specific_text(content, query):
    response = openai.ChatCompletion.create(
        model="gpt-4",
        messages=[
            {"role": "system", "content": "You are an assistant that extracts specific text from large content."},
            {"role": "user", "content": f"Extract text that starts with '{query}' from the following content: \n\n{content}"}
        ]
    )
    return response['choices'][0]['message']['content']

# Example queries
# queries = ["Total Revenue", "Number of Employees"]
queries = ["killed"]

# Extracting specific texts
for i, content in enumerate(webpages_content):
    print(f"Extracted Information from URL {i+1}:")
    for query in queries:
        result = extract_specific_text(content, query)
        print(f"{query}: {result}\n")
```
